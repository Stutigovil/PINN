```python
import tensorflow as tf
import numpy as np

# Convert scaled data to TensorFlow tensors
# X_tf = tf.constant(X_scaled, dtype=tf.float32) # This is now done in the training setup cell
# y_tf = tf.constant(y_scaled, dtype=tf.float32) # This is now done in the training setup cell

# Unscale y_tf to get true outputs in original scale for data loss calculation
# This is important as the physics laws operate on the physical scale
# y_true_unscaled = tf.constant(scaler_y.inverse_transform(y_scaled), dtype=tf.float32) # This should be done per batch in data_loss


# Helper function to interpolate plastic properties
@tf.function
def interpolate_plastic_properties(plastic_strain):
    """Interpolates plastic stress from plastic strain using plastic_properties_df."""
    # Ensure plastic_properties_df is a TensorFlow constant
    plastic_stress_points = tf.constant(plastic_properties_df['Plastic Stress'].values, dtype=tf.float32)
    plastic_strain_points = tf.constant(plastic_properties_df['Plastic Strain'].values, dtype=tf.float32)

    # Ensure the input plastic_strain is within the range of the provided data
    plastic_strain = tf.maximum(0.0, plastic_strain) # Plastic strain cannot be negative
    plastic_strain = tf.minimum(tf.reduce_max(plastic_strain_points), plastic_strain) # Cap at max strain

    # Perform linear interpolation manually using traceable TensorFlow operations
    # Find the indices in plastic_strain_points that would contain each value in plastic_strain
    indices = tf.searchsorted(plastic_strain_points, plastic_strain, side='right')

    # Clamp indices to be within bounds
    # Ensure both arguments to tf.maximum and tf.minimum have the same type
    indices = tf.cast(indices, tf.int64) # Ensure indices is int64 for comparison with size
    indices = tf.maximum(tf.constant(1, dtype=tf.int64), indices) # Cast 1 to int64
    indices = tf.minimum(tf.size(plastic_strain_points, out_type=tf.int64) - 1, indices)
    indices = tf.cast(indices, tf.int32) # Cast back to int32 for gather

    # Get the x and y values for interpolation
    x1 = tf.gather(plastic_strain_points, indices - 1)
    x2 = tf.gather(plastic_strain_points, indices)
    y1 = tf.gather(plastic_stress_points, indices - 1)
    y2 = tf.gather(plastic_stress_points, indices)

    # Perform linear interpolation: y = y1 + (y2 - y1) * (x - x1) / (x2 - x1)
    # Avoid division by zero if x1 equals x2
    delta_x = x2 - x1
    delta_y = y2 - y1
    # Use tf.where to handle cases where delta_x is zero (shouldn't happen with unique points but for safety)
    slope = tf.where(tf.abs(delta_x) < 1e-9, tf.zeros_like(delta_x), delta_y / delta_x)

    interpolated_stress = y1 + slope * (plastic_strain - x1)

    return interpolated_stress


# 1. Data Loss Function (Mean Squared Error for each output branch)
def data_loss(y_true, y_pred):
    """Calculates the data loss between true and predicted outputs for a batch."""
    # Assuming y_pred is a list of tensors corresponding to the model outputs
    # y_true is the true output tensor for the current batch

    # Unscale predicted outputs (which are already for the batch)
    y_pred_combined = tf.concat(y_pred, axis=1)
    # Use TensorFlow operations for unscaling predicted outputs
    y_pred_unscaled = y_pred_combined * tf.constant(scaler_y.scale_, dtype=tf.float32) + tf.constant(scaler_y.mean_, dtype=tf.float32)


    # Unscale the true outputs for the current batch
    # Use TensorFlow operations for unscaling true outputs
    y_true_unscaled = y_true * tf.constant(scaler_y.scale_, dtype=tf.float32) + tf.constant(scaler_y.mean_, dtype=tf.float32)


    # Split the unscaled true and predicted outputs based on the original structure of y
    # This requires knowing the sizes of each output branch
    output_sizes = [6, 6, 12, 1] # Stress, Strain, Displacement, Thickness
    y_true_split = tf.split(y_true_unscaled, output_sizes, axis=1)
    y_pred_split = tf.split(y_pred_unscaled, output_sizes, axis=1)


    loss = 0
    for true_branch, pred_branch in zip(y_true_split, y_pred_split):
        loss += tf.reduce_mean(tf.square(true_branch - pred_branch))

    return loss

# Material properties (example values for Al1050, need to be verified)
YOUNG_MODULUS = 70.0e3  # MPa
POISSON_RATIO = 0.33

# 2. Physics Loss Function Components
def physics_loss(model, X_scaled_tf, plastic_properties_df):
    """Calculates the physics loss based on governing equations for a batch."""
    with tf.GradientTape(persistent=True) as tape:
        # Ensure inputs are being watched for gradients
        tape.watch(X_scaled_tf)

        # Forward pass to get predictions in the scaled space
        stress_pred_scaled, strain_pred_scaled, disp_pred_scaled, thickness_pred_scaled = model(X_scaled_tf)

        # Unscale the predictions to work in the physical space using TensorFlow operations
        y_pred_combined_scaled = tf.concat([stress_pred_scaled, strain_pred_scaled, disp_pred_scaled, thickness_pred_scaled], axis=1)
        y_pred_unscaled = y_pred_combined_scaled * tf.constant(scaler_y.scale_, dtype=tf.float32) + tf.constant(scaler_y.mean_, dtype=tf.float32)


        # Split the unscaled predictions
        output_sizes = [6, 6, 12, 1]
        stress_pred_unscaled, strain_pred_unscaled, disp_pred_unscaled, thickness_pred_unscaled = tf.split(y_pred_unscaled, output_sizes, axis=1)

        # Extract CentroidX, CentroidY, Avg_U3_disp from the unscaled input features using TensorFlow operations
        X_unscaled = X_scaled_tf * tf.constant(scaler_X.scale_, dtype=tf.float32) + tf.constant(scaler_X.mean_, dtype=tf.float32)

        # Ensure that the columns used for spatial gradients are correctly indexed
        # Assuming input features order: ToolDia, WallAngle, InitThickness, CentroidX, CentroidY, Avg_U3_disp
        centroid_x = X_unscaled[:, 3]
        centroid_y = X_unscaled[:, 4]
        avg_u3_disp = X_unscaled[:, 5] # Using Avg_U3_disp as the z-coordinate as per user's instruction


        # Reshape for gradient calculation - ensure they are treated as spatial dimensions
        centroid_x = tf.reshape(centroid_x, (-1, 1))
        centroid_y = tf.reshape(centroid_y, (-1, 1))
        centroid_z_input = tf.reshape(avg_u3_disp, (-1, 1)) # Using Avg_U3_disp as the z-coordinate input

        # Reshape stress and strain for easier access
        S11, S22, S33, S12, S13, S23 = [stress_pred_unscaled[:, i] for i in range(6)]
        E11, E22, E33, E12, E13, E23 = [strain_pred_unscaled[:, i] for i in range(6)]

        # Reshape displacement
        # Assuming the model output for displacement corresponds to average U1, U2, U3 or specific node displacements
        # Based on the 12 outputs, it's likely individual node displacements.
        # Let's calculate average predicted displacements for physics loss if needed, or use a representative node.
        # Given the inputs include Avg_U3_disp and output includes 12 disp values, let's use the predicted displacement outputs directly
        # and relate them to the input spatial coordinates for strain-displacement calculation.
        # For strain-displacement, we need gradients of predicted *displacement components* w.r.t. *spatial coordinates* (CentroidX, CentroidY, Avg_U3_disp).

        # Let's assume the first 3 of the 12 displacement outputs are U1, U2, U3 for a representative point (e.g., centroid)
        # This is a simplification, a more rigorous approach would relate all 12 outputs to the centroid's displacement.
        U1_pred = disp_pred_unscaled[:, 0]
        U2_pred = disp_pred_unscaled[:, 1]
        U3_pred = disp_pred_unscaled[:, 2] # Using the first three predicted displacements as representative U1, U2, U3


    # Equilibrium Loss: âˆ‚Ïƒij/âˆ‚xj = 0
    # Calculate spatial derivatives of stress components with respect to CentroidX, CentroidY, and Avg_U3_disp (as z-input)
    dS11_dx = tape.gradient(S11, centroid_x)
    dS12_dy = tape.gradient(S12, centroid_y)
    dS13_dz = tape.gradient(S13, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    dS21_dx = tape.gradient(S12, centroid_x) # S21 = S12 (symmetric stress tensor)
    dS22_dy = tape.gradient(S22, centroid_y)
    dS23_dz = tape.gradient(S23, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    dS31_dx = tape.gradient(S13, centroid_x) # S31 = S13
    dS32_dy = tape.gradient(S23, centroid_y) # S32 = S23
    dS33_dz = tape.gradient(S33, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    # Handle potential None gradients (if a stress component doesn't depend on a coordinate)
    # Replace None with tf.zeros_like the corresponding stress component
    dS11_dx = tf.zeros_like(S11) if dS11_dx is None else dS11_dx
    dS12_dy = tf.zeros_like(S12) if dS12_dy is None else dS12_dy
    dS13_dz = tf.zeros_like(S13) if dS13_dz is None else dS13_dz

    dS21_dx = tf.zeros_like(S12) if dS21_dx is None else dS21_dx
    dS22_dy = tf.zeros_like(S22) if dS22_dy is None else dS22_dy
    dS23_dz = tf.zeros_like(S23) if dS23_dz is None else dS23_dz

    dS31_dx = tf.zeros_like(S13) if dS31_dx is None else dS31_dx
    dS32_dy = tf.zeros_like(S23) if dS32_dy is None else dS32_dy
    dS33_dz = tf.zeros_like(S33) if dS33_dz is None else dS33_dz


    # Equilibrium equations (ignoring body forces)
    # Ensure shapes are compatible for addition (all should be shape (batch_size, 1) after reshaping coordinates)
    eq_x = dS11_dx + dS12_dy + dS13_dz
    eq_y = dS21_dx + dS22_dy + dS23_dz
    eq_z = dS31_dx + dS32_dy + dS33_dz

    equilibrium_loss = tf.reduce_mean(tf.square(eq_x)) + tf.reduce_mean(tf.square(eq_y)) + tf.reduce_mean(tf.square(eq_z))


    # Strain-Displacement Loss: Îµij = 0.5 * (âˆ‚ui/âˆ‚xj + âˆ‚uj/âˆ‚xi)
    # Calculate spatial derivatives of predicted displacement components with respect to CentroidX, CentroidY, and Avg_U3_disp (as z-input)
    dU1_dx = tape.gradient(U1_pred, centroid_x)
    dU1_dy = tape.gradient(U1_pred, centroid_y)
    dU1_dz = tape.gradient(U1_pred, centroid_z_input)

    dU2_dx = tape.gradient(U2_pred, centroid_x)
    dU2_dy = tape.gradient(U2_pred, centroid_y)
    dU2_dz = tape.gradient(U2_pred, centroid_z_input)

    dU3_dx = tape.gradient(U3_pred, centroid_x)
    dU3_dy = tape.gradient(U3_pred, centroid_y)
    dU3_dz = tape.gradient(U3_pred, centroid_z_input)

    # Handle potential None gradients for displacement derivatives
    dU1_dx = tf.zeros_like(U1_pred) if dU1_dx is None else dU1_dx
    dU1_dy = tf.zeros_like(U1_pred) if dU1_dy is None else dU1_dy
    dU1_dz = tf.zeros_like(U1_pred) if dU1_dz is None else dU1_dz

    dU2_dx = tf.zeros_like(U2_pred) if dU2_dx is None else dU2_dx
    dU2_dy = tf.zeros_like(U2_pred) if dU2_dy is None else dU2_dy
    dU2_dz = tf.zeros_like(U2_pred) if dU2_dz is None else dU2_dz

    dU3_dx = tf.zeros_like(U3_pred) if dU3_dx is None else dU3_dx
    dU3_dy = tf.zeros_like(U3_pred) if dU3_dy is None else dU3_dy
    dU3_dz = tf.zeros_like(U3_pred) if dU3_dz is None else dU3_dz


    # Calculate strain components from displacement gradients
    epsilon_xx = dU1_dx
    epsilon_yy = dU2_dy
    epsilon_zz = dU3_dz

    epsilon_xy = 0.5 * (dU1_dy + dU2_dx)
    epsilon_yz = 0.5 * (dU2_dz + dU3_dy)
    epsilon_zx = 0.5 * (dU3_dx + dU1_dz) # Corrected from epsilon_xz to epsilon_zx


    # Compare with predicted strain components (E11, E22, E33, E12, E13, E23)
    # Reshape predicted strains for comparison - ensure shapes match calculated strains
    pred_E11, pred_E22, pred_E33, pred_E12, pred_E13, pred_E23 = [strain_pred_unscaled[:, i] for i in range(6)]
    pred_E11 = tf.reshape(pred_E11, (-1, 1))
    pred_E22 = tf.reshape(pred_E22, (-1, 1))
    pred_E33 = tf.reshape(pred_E33, (-1, 1))
    pred_E12 = tf.reshape(pred_E12, (-1, 1))
    pred_E13 = tf.reshape(pred_E13, (-1, 1)) # This corresponds to epsilon_zx
    pred_E23 = tf.reshape(pred_E23, (-1, 1)) # This corresponds to epsilon_yz


    strain_disp_loss = tf.reduce_mean(tf.square(epsilon_xx - pred_E11)) + \
                       tf.reduce_mean(tf.square(epsilon_yy - pred_E22)) + \
                       tf.reduce_mean(tf.square(epsilon_zz - pred_E33)) + \
                       tf.reduce_mean(tf.square(epsilon_xy - pred_E12)) + \
                       tf.reduce_mean(tf.square(epsilon_zx - pred_E13)) + \
                       tf.reduce_mean(tf.square(epsilon_yz - pred_E23))


    del tape # Release the tape resources


    # Elastic-Plastic Constitutive Law Loss
    # This is a simplified implementation and might need refinement based on specific material model details.
    # It requires iterative or implicit methods for accurate plasticity modeling, which is complex for a basic PINN.
    # Here, we'll focus on enforcing the stress-strain relationship and yield criterion violation.

    # Calculate elastic strain from predicted stress
    # Assuming isotropic elasticity
    nu = POISSON_RATIO
    E = YOUNG_MODULUS

    # Compliance matrix components (simplified for isotropic)
    # For 3D isotropic elasticity:
    # epsilon_e_ii = (1/E) * (sigma_ii - nu * (sigma_jj + sigma_kk))  (no summation)
    # epsilon_e_ij = (1/G) * sigma_ij = (2*(1+nu)/E) * sigma_ij  (i != j)
    G = E / (2 * (1 + nu)) # Shear modulus

    # Elastic strain components from predicted stress
    epsilon_e_xx = (1/E) * (S11 - nu * (S22 + S33))
    epsilon_e_yy = (1/E) * (S22 - nu * (S11 + S33))
    epsilon_e_zz = (1/E) * (S33 - nu * (S11 + S22))
    epsilon_e_xy = (1/G) * S12
    epsilon_e_xz = (1/G) * S13
    epsilon_e_yz = (1/G) * S23


    # Calculate plastic strain by subtracting elastic strain from total predicted strain
    epsilon_p_xx = E11 - epsilon_e_xx
    epsilon_p_yy = E22 - epsilon_e_yy
    epsilon_p_zz = E33 - epsilon_e_zz
    epsilon_p_xy = E12 - epsilon_e_xy
    epsilon_p_xz = E13 - epsilon_e_xz
    epsilon_p_yz = E23 - epsilon_e_yz

    # Calculate equivalent plastic strain (assuming Von Mises)
    # This is a simplified incremental approach, a full plasticity model is more complex.
    # For a basic PINN, we might focus on the violation of the yield criterion.
    # Equivalent plastic strain: sqrt(2/3 * epsilon_p_ij * epsilon_p_ij) (using tensor notation)
    # For components: sqrt(2/3 * (e_pxx^2 + e_pyy^2 + e_pzz^2 + 2*e_pxy^2 + 2*e_pxz^2 + 2*e_pyz^2))
    equivalent_plastic_strain = tf.sqrt(2/3 * (epsilon_p_xx**2 + epsilon_p_yy**2 + epsilon_p_zz**2 +
                                                2 * epsilon_p_xy**2 + 2 * epsilon_p_xz**2 + 2 * epsilon_p_yz**2))

    # Calculate yield stress from equivalent plastic strain using the plastic properties data (interpolation)
    yield_stress_from_plastic_strain = interpolate_plastic_properties(equivalent_plastic_strain)

    # Calculate Von Mises stress from predicted stress
    von_mises_stress = tf.sqrt(0.5 * ((S11 - S22)**2 + (S22 - S33)**2 + (S33 - S11)**2) +
                               3 * (S12**2 + S13**2 + S23**2))

    # Constitutive Law Loss: Enforce that Von Mises stress <= Yield Stress (or approximately equal for plastic loading)
    # A common approach is to penalize the violation of the yield criterion.
    # Also, penalize if plastic strain is predicted when stress is below initial yield.
    # This is a simplified loss. A more accurate one would involve flow rule and consistency condition.

    # Loss component for violating the yield criterion (only penalize if Von Mises > Yield Stress)
    # Use tf.maximum to create a penalty only when von_mises_stress exceeds yield_stress_from_plastic_strain
    yield_violation_loss = tf.reduce_mean(tf.maximum(0.0, von_mises_stress - yield_stress_from_plastic_strain)**2)

    # Additional loss component: penalize if significant plastic strain is predicted when Von Mises is below initial yield strength
    # Initial yield strength can be taken as the first stress value in plastic_properties_df
    initial_yield_strength = plastic_properties_df['Plastic Stress'].iloc[0] # Assuming the first point is the initial yield stress
    # Penalize if von Mises stress is below initial yield but plastic strain is significant
    below_yield_plastic_strain_loss = tf.reduce_mean(tf.maximum(0.0, initial_yield_strength - von_mises_stress)**2 * equivalent_plastic_strain**2)


    constitutive_loss = yield_violation_loss + below_yield_plastic_strain_loss


    return equilibrium_loss, strain_disp_loss, constitutive_loss

# 3. Total Loss Function (re-defined for completeness)
def total_loss(y_true, y_pred, model, X_scaled_tf, plastic_properties_df, data_weight=1.0, physics_weight=1.0, eq_weight=1.0, sd_weight=1.0, cons_weight=1.0):
    """Calculates the total loss including data and physics components."""
    data_l = data_loss(y_true, y_pred)
    eq_l, sd_l, cons_l = physics_loss(model, X_scaled_tf, plastic_properties_df)

    total_l = data_weight * data_l + physics_weight * (eq_weight * eq_l + sd_weight * sd_l + cons_weight * cons_l)

    # You might want to return individual loss components for monitoring during training
    return total_l, data_l, eq_l, sd_l, cons_l

print("Loss functions updated to handle batch processing and gradient tracking.")

# 6. Create a training loop
# This loop remains the same as it calls the updated total_loss function
for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    total_epoch_loss = 0
    data_epoch_loss = 0
    eq_epoch_loss = 0
    sd_epoch_loss = 0
    cons_epoch_loss = 0
    num_batches = 0

    # Iterate over batches
    for step, (batch_x, batch_y) in enumerate(dataset):
        num_batches += 1
        # 7a. Use tf.GradientTape to record operations
        with tf.GradientTape() as tape:
            # 7b. Perform a forward pass through the model
            # The model has multiple outputs, so y_pred will be a list of tensors
            y_pred = model(batch_x)

            # 7c. Calculate the total loss
            # Pass the true outputs for the batch (batch_y) and the predicted outputs (y_pred)
            # Also pass the model, the scaled input tensor for the batch, and plastic properties
            current_total_loss, current_data_loss, current_eq_loss, current_sd_loss, current_cons_loss = total_loss(
                batch_y, y_pred, model, batch_x, plastic_properties_df
            )

        # 7d. Calculate the gradients
        gradients = tape.gradient(current_total_loss, model.trainable_variables)

        # 7e. Apply the calculated gradients to update the model's weights
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

        total_epoch_loss += current_total_loss.numpy()
        data_epoch_loss += current_data_loss.numpy()
        eq_epoch_loss += current_eq_loss.numpy()
        sd_epoch_loss += current_sd_loss.numpy()
        cons_epoch_loss += current_cons_loss.numpy()

        # Optional: Print loss for the current batch
        # if step % 100 == 0: # Print every 100 batches
        #     print(f"  Batch {step}: Total Loss = {current_total_loss.numpy():.6f}, Data Loss = {current_data_loss.numpy():.6f}, Eq Loss = {current_eq_loss.numpy():.6f}, SD Loss = {current_sd_loss.numpy():.6f}, Cons Loss = {current_cons_loss.numpy():.6f}")


    # 7f. Print the average loss for the epoch
    avg_total_loss = total_epoch_loss / num_batches
    avg_data_loss = data_epoch_loss / num_batches
    avg_eq_loss = eq_epoch_loss / num_batches
    avg_sd_loss = sd_epoch_loss / num_batches
    avg_cons_loss = cons_epoch_loss / num_batches

    print(f"Epoch {epoch+1} Average Loss: Total={avg_total_loss:.6f}, Data={avg_data_loss:.6f}, Equilibrium={avg_eq_loss:.6f}, Strain-Disp={avg_sd_loss:.6f}, Constitutive={avg_cons_loss:.6f}")

print("\nTraining finished.")
```

```python
import tensorflow as tf
import numpy as np

# Helper function to interpolate plastic properties (re-defined for completeness)
@tf.function
def interpolate_plastic_properties(plastic_strain):
    """Interpolates plastic stress from plastic strain using plastic_properties_df."""
    # Ensure plastic_properties_df is a TensorFlow constant
    plastic_stress_points = tf.constant(plastic_properties_df['Plastic Stress'].values, dtype=tf.float32)
    plastic_strain_points = tf.constant(plastic_properties_df['Plastic Strain'].values, dtype=tf.float32)

    # Ensure the input plastic_strain is within the range of the provided data
    plastic_strain = tf.maximum(0.0, plastic_strain) # Plastic strain cannot be negative
    plastic_strain = tf.minimum(tf.reduce_max(plastic_strain_points), plastic_strain) # Cap at max strain

    # Perform linear interpolation manually using traceable TensorFlow operations
    # Find the indices in plastic_strain_points that would contain each value in plastic_strain
    indices = tf.searchsorted(plastic_strain_points, plastic_strain, side='right')

    # Clamp indices to be within bounds
    # Ensure both arguments to tf.maximum and tf.minimum have the same type
    indices = tf.cast(indices, tf.int64) # Ensure indices is int64 for comparison with size
    indices = tf.maximum(tf.constant(1, dtype=tf.int64), indices) # Cast 1 to int64
    indices = tf.minimum(tf.size(plastic_strain_points, out_type=tf.int64) - 1, indices)
    indices = tf.cast(indices, tf.int32) # Cast back to int32 for gather


    # Get the x and y values for interpolation
    x1 = tf.gather(plastic_strain_points, indices - 1)
    x2 = tf.gather(plastic_strain_points, indices)
    y1 = tf.gather(plastic_stress_points, indices - 1)
    y2 = tf.gather(plastic_stress_points, indices)

    # Perform linear interpolation: y = y1 + (y2 - y1) * (x - x1) / (x2 - x1)
    # Avoid division by zero if x1 equals x2
    delta_x = x2 - x1
    delta_y = y2 - y1
    # Use tf.where to handle cases where delta_x is zero (shouldn't happen with unique points but for safety)
    slope = tf.where(tf.abs(delta_x) < 1e-9, tf.zeros_like(delta_x), delta_y / delta_x)

    interpolated_stress = y1 + slope * (plastic_strain - x1)


    return interpolated_stress


# 1. Data Loss Function (Mean Squared Error for each output branch)
def data_loss(y_true, y_pred):
    """Calculates the data loss between true and predicted outputs for a batch."""
    # Assuming y_pred is a list of tensors corresponding to the model outputs
    # y_true is the true output tensor for the current batch

    # Unscale predicted outputs (which are already for the batch) using TensorFlow operations
    y_pred_combined = tf.concat(y_pred, axis=1)
    y_pred_unscaled = y_pred_combined * tf.constant(scaler_y.scale_, dtype=tf.float32) + tf.constant(scaler_y.mean_, dtype=tf.float32)


    # Unscale the true outputs for the current batch using TensorFlow operations
    y_true_unscaled = y_true * tf.constant(scaler_y.scale_, dtype=tf.float32) + tf.constant(scaler_y.mean_, dtype=tf.float32)


    # Split the unscaled true and predicted outputs based on the original structure of y
    # This requires knowing the sizes of each output branch
    output_sizes = [6, 6, 12, 1] # Stress, Strain, Displacement, Thickness
    y_true_split = tf.split(y_true_unscaled, output_sizes, axis=1)
    y_pred_split = tf.split(y_pred_unscaled, output_sizes, axis=1)


    loss = 0
    for true_branch, pred_branch in zip(y_true_split, y_pred_split):
        loss += tf.reduce_mean(tf.square(true_branch - pred_branch))

    return loss

# Material properties (example values for Al1050, need to be verified)
YOUNG_MODULUS = 70.0e3  # MPa
POISSON_RATIO = 0.33

# 2. Physics Loss Function Components (re-defined for completeness)
def physics_loss(model, X_scaled_tf, plastic_properties_df):
    """Calculates the physics loss based on governing equations for a batch."""
    with tf.GradientTape(persistent=True) as tape:
        # Ensure inputs are being watched for gradients
        tape.watch(X_scaled_tf)

        # Forward pass to get predictions in the scaled space
        stress_pred_scaled, strain_pred_scaled, disp_pred_scaled, thickness_pred_scaled = model(X_scaled_tf)

        # Unscale the predictions to work in the physical space using TensorFlow operations
        y_pred_combined_scaled = tf.concat([stress_pred_scaled, strain_pred_scaled, disp_pred_scaled, thickness_pred_scaled], axis=1)
        # Convert scaler attributes to tensors for unscaling
        scaler_y_scale_tf = tf.constant(scaler_y.scale_, dtype=tf.float32)
        scaler_y_mean_tf = tf.constant(scaler_y.mean_, dtype=tf.float32)
        y_pred_unscaled = y_pred_combined_scaled * scaler_y_scale_tf + scaler_y_mean_tf


        # Split the unscaled predictions
        output_sizes = [6, 6, 12, 1]
        stress_pred_unscaled, strain_pred_unscaled, disp_pred_unscaled, thickness_pred_unscaled = tf.split(y_pred_unscaled, output_sizes, axis=1)

        # Extract CentroidX, CentroidY from the unscaled input features using TensorFlow operations
        # Convert scaler attributes to tensors for unscaling X
        scaler_X_scale_tf = tf.constant(scaler_X.scale_, dtype=tf.float32)
        scaler_X_mean_tf = tf.constant(scaler_X.mean_, dtype=tf.float32)
        X_unscaled = X_scaled_tf * scaler_X_scale_tf + scaler_X_mean_tf


        # Ensure that the columns used for spatial gradients are correctly indexed
        # Assuming input features order: ToolDia, WallAngle, InitThickness, CentroidX, CentroidY, Avg_U3_disp
        centroid_x = X_unscaled[:, 3]
        centroid_y = X_unscaled[:, 4]
        centroid_z_input = X_unscaled[:, 5] # Using Avg_U3_disp as the z-coordinate input


        # Ensure centroid coordinates are watched by the tape for spatial gradients
        tape.watch(centroid_x)
        tape.watch(centroid_y)
        tape.watch(centroid_z_input)


        # Reshape CentroidX, CentroidY, CentroidZ for gradient calculation
        centroid_x = tf.reshape(centroid_x, (-1, 1))
        centroid_y = tf.reshape(centroid_y, (-1, 1))
        centroid_z = tf.reshape(centroid_z_input, (-1, 1))


        # Reshape stress and strain for easier access
        S11, S22, S33, S12, S13, S23 = [stress_pred_unscaled[:, i] for i in range(6)]
        E11, E22, E33, E12, E13, E23 = [strain_pred_unscaled[:, i] for i in range(6)]

        # Reshape displacement
        # Assuming the model output for displacement corresponds to average U1, U2, U3 or specific node displacements
        # Based on the 12 outputs, it's likely individual node displacements.
        # Let's calculate average predicted displacements for physics loss if needed, or use a representative node.
        # Given the inputs include Avg_U3_disp and output includes 12 disp values, let's use the predicted displacement outputs directly
        # and relate them to the input spatial coordinates for strain-displacement calculation.
        # For strain-displacement, we need gradients of predicted *displacement components* w.r.t. *spatial coordinates* (CentroidX, CentroidY, Avg_U3_disp).

        # Let's assume the first 3 of the 12 displacement outputs are U1, U2, U3 for a representative point (e.g., centroid)
        # This is a simplification, a more rigorous approach would relate all 12 outputs to the centroid's displacement.
        U1_pred = disp_pred_unscaled[:, 0]
        U2_pred = disp_pred_unscaled[:, 1]
        U3_pred = disp_pred_unscaled[:, 2] # Using the first three predicted displacements as representative U1, U2, U3


    # Equilibrium Loss: âˆ‚Ïƒij/âˆ‚xj = 0
    # Calculate spatial derivatives of stress components with respect to CentroidX, CentroidY, and Avg_U3_disp (as z-input)
    dS11_dx = tape.gradient(S11, centroid_x)
    dS12_dy = tape.gradient(S12, centroid_y)
    dS13_dz = tape.gradient(S13, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    dS21_dx = tape.gradient(S12, centroid_x) # S21 = S12 (symmetric stress tensor)
    dS22_dy = tape.gradient(S22, centroid_y)
    dS23_dz = tape.gradient(S23, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    dS31_dx = tape.gradient(S13, centroid_x) # S31 = S13
    dS32_dy = tape.gradient(S23, centroid_y) # S32 = S23
    dS33_dz = tape.gradient(S33, centroid_z_input) # Gradient w.r.t. Avg_U3_disp as z-input

    # Handle potential None gradients (if a stress component doesn't depend on a coordinate)
    # Replace None with tf.zeros_like the corresponding stress component
    dS11_dx = tf.zeros_like(S11) if dS11_dx is None else dS11_dx
    dS12_dy = tf.zeros_like(S12) if dS12_dy is None else dS12_dy
    dS13_dz = tf.zeros_like(S13) if dS13_dz is None else dS13_dz

    dS21_dx = tf.zeros_like(S12) if dS21_dx is None else dS21_dx
    dS22_dy = tf.zeros_like(S22) if dS22_dy is None else dS22_dy
    dS23_dz = tf.zeros_like(S23) if dS23_dz is None else dS23_dz

    dS31_dx = tf.zeros_like(S13) if dS31_dx is None else dS31_dx
    dS32_dy = tf.zeros_like(S23) if dS32_dy is None else dS32_dy
    dS33_dz = tf.zeros_like(S33) if dS33_dz is None else dS33_dz


    # Equilibrium equations (ignoring body forces)
    # Ensure shapes are compatible for addition (all should be shape (batch_size, 1) after reshaping coordinates)
    eq_x = dS11_dx + dS12_dy + dS13_dz
    eq_y = dS21_dx + dS22_dy + dS23_dz
    eq_z = dS31_dx + dS32_dy + dS33_dz

    equilibrium_loss = tf.reduce_mean(tf.square(eq_x)) + tf.reduce_mean(tf.square(eq_y)) + tf.reduce_mean(tf.square(eq_z))


    # Strain-Displacement Loss: Îµij = 0.5 * (âˆ‚ui/âˆ‚xj + âˆ‚uj/âˆ‚xi)
    # Calculate spatial derivatives of predicted displacement components with respect to CentroidX, CentroidY, and Avg_U3_disp (as z-input)
    with tf.GradientTape(persistent=True) as strain_disp_tape:
         strain_disp_tape.watch(X_scaled_tf) # Watch the input to the model

         # Re-calculate unscaled predictions within a new tape for strain-displacement
         stress_pred_scaled_sd, strain_pred_scaled_sd, disp_pred_scaled_sd, thickness_pred_scaled_sd = model(X_scaled_tf)

         y_pred_combined_scaled_sd = tf.concat([stress_pred_scaled_sd, strain_pred_scaled_sd, disp_pred_scaled_sd, thickness_pred_scaled_sd], axis=1)
         # Convert scaler attributes to tensors for unscaling
         scaler_y_scale_tf_sd = tf.constant(scaler_y.scale_, dtype=tf.float32)
         scaler_y_mean_tf_sd = tf.constant(scaler_y.mean_, dtype=tf.float32)
         y_pred_unscaled_sd = y_pred_combined_scaled_sd * scaler_y_scale_tf_sd + scaler_y_mean_tf_sd


         # Split the unscaled predictions for strain-displacement calculation
         output_sizes = [6, 6, 12, 1]
         stress_pred_unscaled_sd, strain_pred_unscaled_sd, disp_pred_unscaled_sd, thickness_pred_unscaled_sd = tf.split(y_pred_unscaled_sd, output_sizes, axis=1)

         U1_sd = disp_pred_unscaled_sd[:, 0] # Using the first node's U1 as representative
         U2_sd = disp_pred_unscaled_sd[:, 1] # Using the first node's U2 as representative
         U3_sd = disp_pred_unscaled_sd[:, 2] # Using the first node's U3 as representative

         # Extract CentroidX, CentroidY, and Avg_U3_disp from the unscaled input features within this tape's scope
         # Convert scaler attributes to tensors for unscaling X
         scaler_X_scale_tf_sd = tf.constant(scaler_X.scale_, dtype=tf.float32)
         scaler_X_mean_tf_sd = tf.constant(scaler_X.mean_, dtype=tf.float32)
         X_unscaled_sd = X_scaled_tf * scaler_X_scale_tf_sd + scaler_X_mean_tf_sd

         centroid_x_sd = X_unscaled_sd[:, 3]
         centroid_y_sd = X_unscaled_sd[:, 4]
         centroid_z_sd = X_unscaled_sd[:, 5] # Using Avg_U3_disp as the z-coordinate

         # Ensure centroid coordinates are watched by the tape for spatial gradients
         strain_disp_tape.watch(centroid_x_sd)
         strain_disp_tape.watch(centroid_y_sd)
         strain_disp_tape.watch(centroid_z_sd)

         # Reshape for gradient calculation
         centroid_x_sd = tf.reshape(centroid_x_sd, (-1, 1))
         centroid_y_sd = tf.reshape(centroid_y_sd, (-1, 1))
         centroid_z_sd = tf.reshape(centroid_z_sd, (-1, 1))


         # Calculate displacement gradients
         dU1_dx = strain_disp_tape.gradient(U1_sd, centroid_x_sd)
         dU1_dy = strain_disp_tape.gradient(U1_sd, centroid_y_sd)
         dU1_dz = strain_disp_tape.gradient(U1_sd, centroid_z_sd) # Gradient w.r.t. Avg_U3_disp as z-input

         dU2_dx = strain_disp_tape.gradient(U2_sd, centroid_x_sd)
         dU2_dy = strain_disp_tape.gradient(U2_sd, centroid_y_sd)
         dU2_dz = strain_disp_tape.gradient(U2_sd, centroid_z_sd) # Gradient w.r.t. Avg_U3_disp as z-input

         dU3_dx = strain_disp_tape.gradient(U3_sd, centroid_x_sd)
         dU3_dy = strain_disp_tape.gradient(U3_sd, centroid_y_sd)
         dU3_dz = strain_disp_tape.gradient(U3_sd, centroid_z_sd) # Gradient w.r.t. Avg_U3_disp as z-input

         # Handle potential None gradients for displacement derivatives
         dU1_dx = tf.zeros_like(U1_sd) if dU1_dx is None else dU1_dx
         dU1_dy = tf.zeros_like(U1_sd) if dU1_dy is None else dU1_dy
         dU1_dz = tf.zeros_like(U1_sd) if dU1_dz is None else dU1_dz

         dU2_dx = tf.zeros_like(U2_sd) if dU2_dx is None else dU2_dx
         dU2_dy = tf.zeros_like(U2_sd) if dU2_dy is None else dU2_dy
         dU2_dz = tf.zeros_like(U2_sd) if dU2_dz is None else dU2_dz

         dU3_dx = tf.zeros_like(U3_sd) if dU3_dx is None else dU3_dx
         dU3_dy = tf.zeros_like(U3_sd) if dU3_dy is None else dU3_dy
         dU3_dz = tf.zeros_like(U3_sd) if dU3_dz is None else dU3_dz


         # Calculate strain components from displacement gradients
         epsilon_xx = dU1_dx
         epsilon_yy = dU2_dy
         epsilon_zz = dU3_dz # Assuming dU3/dZ represents epsilon_zz

         epsilon_xy = 0.5 * (dU1_dy + dU2_dx)
         epsilon_xz = 0.5 * (dU1_dz + dU3_dx) # Assuming dU1/dZ and dU3/dX are relevant
         epsilon_yz = 0.5 * (dU2_dz + dU3_dy) # Assuming dU2/dZ and dU3/dY are relevant

         # Compare with predicted strain components
         # Reshape predicted strains for comparison
         pred_E11, pred_E22, pred_E33, pred_E12, pred_E13, pred_E23 = tf.split(strain_pred_unscaled_sd, 6, axis=1)
         pred_E11 = tf.reshape(pred_E11, (-1, 1))
         pred_E22 = tf.reshape(pred_E22, (-1, 1))
         pred_E33 = tf.reshape(pred_E33, (-1, 1))
         pred_E12 = tf.reshape(pred_E12, (-1, 1))
         pred_E13 = tf.reshape(pred_E13, (-1, 1))
         pred_E23 = tf.reshape(pred_E23, (-1, 1))


         strain_disp_loss = tf.reduce_mean(tf.square(epsilon_xx - pred_E11)) + \
                            tf.reduce_mean(tf.square(epsilon_yy - pred_E22)) + \
                            tf.reduce_mean(tf.square(epsilon_zz - pred_E33)) + \
                            tf.reduce_mean(tf.square(epsilon_xy - pred_E12)) + \
                            tf.reduce_mean(tf.square(epsilon_xz - pred_E13)) + \
                            tf.reduce_mean(tf.square(epsilon_yz - pred_E23))

    del strain_disp_tape # Release the tape resources


    # Elastic-Plastic Constitutive Law Loss
    # This is a simplified implementation and might need refinement based on specific material model details.
    # It requires iterative or implicit methods for accurate plasticity modeling, which is complex for a basic PINN.
    # Here, we'll focus on enforcing the stress-strain relationship and yield criterion violation.

    # Calculate elastic strain from predicted stress
    # Assuming isotropic elasticity
    nu = POISSON_RATIO
    E = YOUNG_MODULUS

    # Compliance matrix components (simplified for isotropic)
    C11 = 1/E
    C12 = -nu/E
    # Shear modulus (G = E / (2*(1+nu)))
    G = E / (2 * (1 + nu))
    C44 = 1/G


    # Elastic strain components from predicted stress
    epsilon_e_xx = C11 * S11 + C12 * S22 + C12 * S33
    epsilon_e_yy = C12 * S11 + C11 * S22 + C12 * S33
    epsilon_e_zz = C12 * S11 + C12 * S22 + C11 * S33
    epsilon_e_xy = C44 * S12 # Simplified shear
    epsilon_e_xz = C44 * S13 # Simplified shear
    epsilon_e_yz = C44 * S23 # Simplified shear


    # Calculate plastic strain by subtracting elastic strain from total predicted strain
    epsilon_p_xx = E11 - epsilon_e_xx
    epsilon_p_yy = E22 - epsilon_e_yy
    epsilon_p_zz = E33 - epsilon_e_zz
    epsilon_p_xy = E12 - epsilon_e_xy
    epsilon_p_xz = E13 - epsilon_e_xz
    epsilon_p_yz = E23 - epsilon_e_yz

    # Calculate equivalent plastic strain (assuming Von Mises)
    # This is a simplified incremental approach, a full plasticity model is more complex.
    # For a basic PINN, we might focus on the violation of the yield criterion.
    # Equivalent plastic strain increment: d(epsilon_p_eq) = sqrt(2/3 * d(epsilon_p_ij) * d(epsilon_p_ij))
    # Total equivalent plastic strain: epsilon_p_eq = integral(d(epsilon_p_eq))
    # This requires integrating plastic strain increments over the loading history, which is hard in a static PINN.
    # A simpler approach is to use the current plastic strain state.
    equivalent_plastic_strain = tf.sqrt(2/3 * (epsilon_p_xx**2 + epsilon_p_yy**2 + epsilon_p_zz**2 +
                                                2 * epsilon_p_xy**2 + 2 * epsilon_p_xz**2 + 2 * epsilon_p_yz**2))

    # Calculate yield stress from equivalent plastic strain using the plastic properties data
    yield_stress_from_plastic_strain = interpolate_plastic_properties(equivalent_plastic_strain)

    # Calculate Von Mises stress from predicted stress
    von_mises_stress = tf.sqrt(0.5 * ((S11 - S22)**2 + (S22 - S33)**2 + (S33 - S11)**2) +
                               3 * (S12**2 + S13**2 + S23**2))

    # Constitutive Law Loss: Enforce that Von Mises stress <= Yield Stress (or approximately equal for plastic loading)
    # A common approach is to penalize the violation of the yield criterion.
    # Also, penalize if plastic strain is predicted when stress is below yield.
    # This is a simplified loss. A more accurate one would involve flow rule and consistency condition.

    # Loss component for violating the yield criterion (only penalize if Von Mises > Yield Stress)
    yield_violation_loss = tf.reduce_mean(tf.maximum(0.0, von_mises_stress - yield_stress_from_plastic_strain)**2)

    # Additional loss component: penalize if significant plastic strain is predicted when Von Mises is below initial yield strength
    # Initial yield strength can be taken as the first stress value in plastic_properties_df
    initial_yield_strength = plastic_properties_df['Plastic Stress'].iloc[0]
    below_yield_plastic_strain_loss = tf.reduce_mean(tf.maximum(0.0, initial_yield_strength - von_mises_stress)**2 * equivalent_plastic_strain**2)


    constitutive_loss = yield_violation_loss + below_yield_plastic_strain_loss

    # Don't forget to delete the tapes
    del tape


    return equilibrium_loss, strain_disp_loss, constitutive_loss

# 3. Total Loss Function (re-defined for completeness)
def total_loss(y_true, y_pred, model, X_scaled_tf, plastic_properties_df, data_weight=1.0, physics_weight=1.0, eq_weight=1.0, sd_weight=1.0, cons_weight=1.0):
    """Calculates the total loss including data and physics components."""
    data_l = data_loss(y_true, y_pred)
    eq_l, sd_l, cons_l = physics_loss(model, X_scaled_tf, plastic_properties_df)

    total_l = data_weight * data_l + physics_weight * (eq_weight * eq_l + sd_weight * sd_l + cons_weight * cons_l)

    # You might want to return individual loss components for monitoring during training
    return total_l, data_l, eq_l, sd_l, cons_l


# 6. Create a training loop
for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    total_epoch_loss = 0
    data_epoch_loss = 0
    eq_epoch_loss = 0
    sd_epoch_loss = 0
    cons_epoch_loss = 0
    num_batches = 0

    # Iterate over batches
    for step, (batch_x, batch_y) in enumerate(dataset):
        num_batches += 1
        # 7a. Use tf.GradientTape to record operations
        with tf.GradientTape() as tape:
            # 7b. Perform a forward pass through the model
            # The model has multiple outputs, so y_pred will be a list of tensors
            y_pred = model(batch_x)

            # 7c. Calculate the total loss
            # Pass the true outputs for the batch (batch_y) and the predicted outputs (y_pred)
            # Also pass the model, the scaled input tensor for the batch, and plastic properties
            current_total_loss, current_data_loss, current_eq_loss, current_sd_loss, current_cons_loss = total_loss(
                batch_y, y_pred, model, batch_x, plastic_properties_df
            )

        # 7d. Calculate the gradients
        gradients = tape.gradient(current_total_loss, model.trainable_variables)

        # 7e. Apply the calculated gradients to update the model's weights
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))

        total_epoch_loss += current_total_loss.numpy()
        data_epoch_loss += current_data_loss.numpy()
        eq_epoch_loss += current_eq_loss.numpy()
        sd_epoch_loss += current_sd_loss.numpy()
        cons_epoch_loss += current_cons_loss.numpy()

        # Optional: Print loss for the current batch
        # if step % 100 == 0: # Print every 100 batches
        #     print(f"  Batch {step}: Total Loss = {current_total_loss.numpy():.6f}, Data Loss = {current_data_loss.numpy():.6f}, Eq Loss = {current_eq_loss.numpy():.6f}, SD Loss = {current_sd_loss.numpy():.6f}, Cons Loss = {current_cons_loss.numpy():.6f}")


    # 7f. Print the average loss for the epoch
    avg_total_loss = total_epoch_loss / num_batches
    avg_data_loss = data_epoch_loss / num_batches
    avg_eq_loss = eq_epoch_loss / num_batches
    avg_sd_loss = sd_epoch_loss / num_batches
    avg_cons_loss = cons_epoch_loss / num_batches

    print(f"Epoch {epoch+1} Average Loss: Total={avg_total_loss:.6f}, Data={avg_data_loss:.6f}, Equilibrium={avg_eq_loss:.6f}, Strain-Disp={avg_sd_loss:.6f}, Constitutive={avg_cons_loss:.6f}")

print("\nTraining finished.")
```
